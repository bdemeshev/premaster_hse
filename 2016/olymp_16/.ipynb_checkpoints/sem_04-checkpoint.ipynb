#### Maximum likelihood - Метод максимального правдоподобия

идея ML: выбирать такую, $\hat\theta$, оценку неизвестного параметра $\theta$, при которой вероятность получить фактические данные максимальна

То, что мы наблюдаем --- это часто встречающееся событие

пример с фонтаном. 
* А: Фонтан здесь включают каждый день
* B: Фонтан включают раз в десять лет


##### Задачка. 

Каждый день Кот Матроскин ходит ловить рыбу.

* Щука - вероятность $p$
* Карась - вероятность $2p$
* Сом - вероятность $1-3p$

Неизвестный параметр: $p$

Фактические данные: $y_1 =\text{Щука}$, $y_2=\text{Сом}$, $y_3 =\text{Щука}$, $y_4=\text{Карась}$.

Отдельные наблюдения независимы (independent)

Estimate $p$ using ML

##### Решение:

Мы найдем не истинную $p$, а только оценку (estimate), $\hat p_{ML}$

$$
\mathcal{L}=P(y_1 =\text{Щука}, y_2=\text{Сом}, y_3 =\text{Щука}, y_4=\text{Карась}) =
P(y_1 =\text{Щука})\cdot P( y_2=\text{Сом})\cdot P( y_3 =\text{Щука})\cdot P( y_4=\text{Карась}) = p\cdot (1-3p)\cdot p\cdot 2p
=2p^3 (1-3p)$$

Функция правдоподобия (likelihood function)

$$
\ln \mathcal{L} = \ln 2p^3 (1-3p) = \ln 2 + \ln p^3 + \ln (1-3p) = \ln 2 + 3\ln p + \ln (1-3p)
$$

Максимизируем:

$$
\ln \mathcal{L}' = 3/p - 3/(1-3p)
$$

Предположим, что нашие данные - это самые часто встречающиеся!

$$
3/\hat p - 3/(1-3\hat p)=0
$$

Ответ: $\hat p_{ML}=1/4$

##### Интерпретация logit-probit



$y_i$ - 1, если клиент вернул кредит, и 0, если не вернул

Существует некая "склонность", $y_i^*$, вернуть кредит (не вероятность, принимает любые значения от минус до плюс бесконечности). Склонность строго связана с вероятностью.

$$
y_i^* = \beta_0 + \beta_1 age_i + \beta_2 male_i + u_i
$$

Видимый нами $y_i$ определяется через склонностью $y_i^*$ бинарно:

* Если склонность $y_i^*>0$, то клиент возвращает кредит, $y_i = 1$
* Если склонность $y_i^*<0$, то клиент не возвращает кредит, $y_i =0$


* Probit: $u_i \sim N(0;1)$
* Logit: $u_i \sim Logis \approx N(0; 3.3)$

```r
qplot(x = u, y = pnorm(u/sqrt(3.3)))
```

##### интерпретация коэффициентов logit модели

* probability / odds
* вероятность / шансы

например:
* вероятность равна 0.2
* шансы 1:4


logit:
* вероятность $P(y_i = 1)=F(\beta_0 + \beta_1 age_i + \beta_2 male_i)$
* шансы $P(y_i = 1)/(1-P(y_i = 1)) = ... = \exp(\beta_0 + \beta_1 age_i + \beta_2 male_i)$

Шансы в logit-модели. Берём логарифм:
$$
\ln (\text{Шансы на возврат кредита}) = \beta_0 + \beta_1 age_i + \beta_2 male_i
$$

Две количественных меры зависимости:
* производная спроса по доходу равна $-2$: Если $x$ растёт на 1 единицу, то $y$ падает на 2 единицы.
* эластичность спроса по доходу равна $-2$: Если $x$ растёт на 1 процент, то $y$ падает на 2 процента.

$$
\text{Elasticity} = \frac{dy/y}{dx/x}=\frac{d\ln y}{d\ln x}
$$

Вернёмся к нашей логит модели:
$$
\beta_1 = \frac{d\ln \text{Шансы}}{d age}=\frac{d \text{Шансы}/\text{ Шансы}}{d age}
$$

Пример интерпретации:

Оценили с помощью ML коэффициенты в логит модели и оказалось, что $\hat\beta_1 = 0.08$

При росте возраста на 1 год шансы на возврат кредита растут на 8%

На языке вероятностей тоже можно посчитать предельный эффект (marginal effect):

$$
\frac{d P(y_i =1)}{d age_i} = F'(\beta_0 + \beta_1 age_i + \beta_2 male_i)\cdot \beta_1
$$

Следствие формулы для предельного эффекта в логит-модели!

В правой части формулы есть значения $age_i$ и $male_i$, это означает, что предельный эффект увеличения возраста на вероятность возврата кредита зависит от $age_i$ и $male_i$. То есть предельный эффект разный для разных инидивидов!
